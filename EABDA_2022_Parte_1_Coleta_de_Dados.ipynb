## Coletando todas as páginas do Evento

```python
# Criando uma função que contém todos os passos descritos anteriormente
# No final do método, é gerado um dicionário contendo os dados extraídos
def extract_info(url):
  response = requests.get(url)

  conteudo_html = response.content.decode('utf-8')

  soup = BeautifulSoup(conteudo_html, 'html.parser')

  title = soup.find('h5', {'class' : 'page__title-3'}).text

  nome_ministrante = soup.find('div', {'class' : 'course__teacher-info-3'}).h4.text

  paragrafos_resumo = soup.find('div', {'class' : 'course__description'}).findAll('p')
  resumo = ' '.join([paragrafo.text.strip() for paragrafo in paragrafos_resumo])

  paragrafos_resumo = soup.find('div', {'class' : 'course__description'}).findAll('p')
  resumo = ' '.join([paragrafo.text.strip() for paragrafo in paragrafos_resumo])

  price_features = soup.find('div', {'class' : 'price__features'}).ul.findAll('li')

  data_hora = price_features[0].text
  data_hora

  data = re.findall(r'\d{1,2}/\d{1,2}/\d{4}', data_hora)[0]

  try:
    hora = re.findall(r'\d{1,2}:\d{1,2} - \d{1,2}:\d{1,2}', data_hora)[0]
  except:
    hora = re.findall(r'(\d{1,2}h-\d{1,2}h\d{1,2})', data_hora)[0]


  try:
    modalidade = price_features[1].text
    modalidade = re.findall(r'Modalidade: (.+) \(Clique Aqui\)', modalidade)[0]
  except:
    modalidade = ''

  try:
    carga_horaria = price_features[2].text
    carga_horaria = re.findall(r'(\d+) horas', carga_horaria)[0]
  except:
    carga_horaria = '-'

  bibliografia =  [bibliografia.text for bibliografia in soup.find('div', {'class' : 'course__description-list'}).findAll('li')]

  curriculum_ministrante = ' '.join([paragraph.text.strip() for paragraph in soup.find('div', {'class' : 'course__curriculum'}).findAll('p')] )

  link_imagem_ministrante = soup.find('div', {'class' : 'course__teacher-thumb-3'}).img['src']
  nome_imagem = os.path.basename(link_imagem_ministrante)

  image_content = requests.get("{}/{}".format(os.path.dirname(URL),link_imagem_ministrante))
  with open(nome_imagem, 'wb') as file:
    file.write(image_content.content)

  data = {
      'título' : title,
      'nome_ministrante' : nome_ministrante,
      'resumo' : resumo,
      'modalidade' : modalidade,
      'data' : data,
      'horario' : hora,
      'carga_horaria' : carga_horaria,
      'bibliografia' : bibliografia,
      'curriculum_ministrante' : curriculum_ministrante,
      'path_imagem' : nome_imagem
  }

  return data
```

## Crawler do Site Canal Rural

```python
# To add a new cell, type '# %%'
# To add a new markdown cell, type '# %% [markdown]'
# %% [markdown]
# # Imports

# %%
from bs4 import BeautifulSoup
from urllib.request import Request, urlopen
import pandas as pd
import os
import sys
import requests
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:88.0) Gecko/20100101 Firefox/88.0'}
commodity = 'biocombustivel'
main_url = 'https://www.noticiasagricolas.com.br/noticias'
dir_base = '/media/...' # Defina aqui o diretório onde irá salvar as páginas


session = requests.Session()
session.headers.update(headers)
retry = Retry(connect=100, backoff_factor=3)
adapter = HTTPAdapter(max_retries=retry)
session.mount('http://', adapter)
session.mount('https://', adapter)


def trata_html(input):
    return " ".join(input.split()).replace('> <', '><')


def salva_pagina(path, conteudo):
    with open(path, 'w') as arquivo:
        arquivo.write(conteudo)


def requisicao(url):
    print(url)
    response = session.get(url)
    # home_html contém a página principal (esta página já está sendo decodificada para utf-8 e um processo para reduzir número de espaços no texto está sendo aplicado)
    pagina = trata_html(response.content.decode('utf-8'))
    return pagina

def crawler():
    home_url = f'{main_url}/{commodity}'
    home_html = requisicao(home_url)

    # Convertendo a página inicial coletada para um Objeto BeautifulSoup
    home_soup = BeautifulSoup(home_html, 'html.parser')
    ul_paginacao = home_soup.find('ul', {'class' : 'paginacao'})

    # Obtendo o índice da última página
    ultima_pagina = int(ul_paginacao.findAll('li', {'class':'page-number'})[-1].get_text())
    print(f'Páginas 1 ... {ultima_pagina}')

    for num_paginacao in range(1, ultima_pagina + 1):
        url = f'{home_url}/?page={num_paginacao}'
        print('============================================================')
        print(url)
        print('============================================================')
        paginacao_atual = requisicao(url)

        # Convertendo a página inicial coletada para um Objeto BeautifulSoup
        soup = BeautifulSoup(paginacao_atual, 'html.parser')

        #Extraindo todos os links de notícias da página atual
        li_links = soup.findAll('li', {'class':['horizontal', 'com-hora']})
        for link in li_links:
            url_a = link.find('a').attrs['href'].split('/')[-1]
            url_noticia = f'{home_url}/{url_a}'
            url_file = url_noticia.replace('/', '___')
            try:
                pagina_noticia = requisicao(url_noticia)
                path = os.path.join(dir_base, commodity, url_file)
                par_dir = os.path.dirname(path)
                if not os.path.exists(par_dir):
                    os.makedirs(par_dir)
                salva_pagina(path, pagina_noticia)
            except:
                pass

    print('DONE!')

```

## Scrapper do Site Canal Rural

```python

import pandas as pd
import os
import re
from bs4 import BeautifulSoup

# Definindo os caminhos do diretório de entrada e de saída
dir_input = '/media/...'
dir_output = '/media/...'

# Expressões regulares
regex_data_hora = re.compile(r'(\d{1,2}/\d{1,2}/\d{4}) ÀS (\d{1,2}h\d{1,2})')
regex_fonte = re.compile(r'Por (.+?) +-')

# Funcao para extrair as informações da notícia
def extrair_informacoes(conteudo):
    soup = BeautifulSoup(conteudo, 'html.parser')

    principal = soup.main

    if principal == None:
        return None

    entrada = {}

    #Extraindo o tema da página
    entrada['tema'] = principal.find('div', {'class' : 'chapeu'}).get_text().strip()

    #Extranindo o título da página
    entrada['titulo'] = principal.h1.get_text().strip()

    #Extraindo o resumo da página
    entrada['resumo'] = principal.find('p', {'class': 'resumo'}).get_text().strip()

    #Extraindo a data e hora
    data_hora = regex_data_hora.findall(principal.find('p', {'class': 'data-autor'}).find('span').get_text().strip())
    if len(data_hora) > 0:
        data_hora = data_hora[0]
        entrada['data'] = data_hora[0]
        entrada['hora'] = data_hora[1].replace('h',':')

    # Extraindo o conteudo da matéria
    div_texto_noticia = principal.find('div', {'class' : 'texto-noticia'})
    entrada['conteudo'] = '\n'.join([paragrafo.get_text().strip() for paragrafo in div_texto_noticia.findAll('p', {'class':''}) if not str(paragrafo).startswith('<p><a')])

    #Extraindo a fonte da matéria
    fonte = regex_fonte.findall(principal.find('p', {'class': 'data-autor'}).get_text())
    if len(fonte) > 0:
        entrada['fonte'] = fonte[0]


    return entrada

# Listando os diretórios do diretório base (cada diretório base irá virar um .csv na pasta de saída)
diretorios = os.listdir(dir_input)

for diretorio in diretorios:
    current_dir = os.path.join(dir_input, diretorio)

    if not os.path.isdir(current_dir):
        continue

    path_output = os.path.join(dir_output, f'{diretorio}.csv')
    if os.path.exists(path_output):
        continue

    #Criando o DataFrame para armazenar os dados
    df = pd.DataFrame()
    #Listando os arquivos do diretorio
    files = os.listdir(current_dir)
    for file in files:

        # Lendo o arquivo e criando um objeto do BeautifulSoup
        conteudo = ''
        path_file = os.path.join(current_dir, file)
        print(path_file)
        with open(path_file) as noticia:
            conteudo = ''.join(noticia.readlines())
            conteudo = re.sub('\n{2,}','\n', conteudo).strip()

            #Extraindo a url a partir no nome do arquivo
        entrada = extrair_informacoes(conteudo)
        if entrada == None:
            continue

        entrada['url'] = file.replace('___', '/')

        #Concatenando a entrada no DataFrame
        df = df.append(entrada, ignore_index=True)

    #Gravando o CSV de saída

    df.to_csv(path_output, index=None)

print('DONE!')
```
